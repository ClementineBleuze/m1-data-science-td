# Statistics and Visualisation

### Calculating mean, median and mode

```python
import numpy as np
from scipy import stats
sentence_lengths = [2,2,4,6,8]
print("Mean: ", np.mean(sentence_lengths))
print("Median: ",np.median(sentence_lengths))
print("Mode: ",stats.mode(sentence_lengths))
```

### Calculating the range

```python
import numpy as np
sentence_lengths = [15, 16, 16, 17, 18, 19, 30, 110]
np.ptp(sentence_lengths)
```

### Calculating variance and standard deviation

```python
import numpy as np
l = [4, 8, 10, 11, 18, 19, 20, 21, 23, 24, 50]
print("Variance: ",np.var(l))
print("Standard Deviation: ",np.std(l))
```

### Calculating Q1, Q2 and Q3

```python
import numpy as np
l = [2, 4, 8, 9, 10, 11, 12, 13, 15, 20, 22] 
#l = [4, 8, 10, 11, 18, 19, 20, 21, 23, 24, 50]
print("Q1: ",np.percentile(l,25))
print("Q2: ",np.percentile(l,50))
print("Q3: ",np.percentile(l,75))
```

### Calculating the inter quartile range

```python
from scipy.stats import iqr
print("IQR: ",iqr(l))
```

### Plotting the frequency distribution

```python
%matplotlib inline
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
# Store movies data into Pandas dataframe
movies_df = pd.read_csv("data/movies_500.csv", sep = "\t")
# Concatenate values of 'story_str' columns
reviews = movies_df.story_str.str.cat(sep=' ')
# Tokenize reviews
tokens = word_tokenize(reviews)
# Get set of distinct tokens present in reviews
vocabulary = set(tokens)
# Get Frequency distribution
frequency_dist = nltk.FreqDist(tokens)
# Plot the frequency distribution for the 10 most
# frequent words
frequency_dist.plot(10)
```

### Chi-square and P-value

```python
import numpy as np
from scipy.stats import chi2_contingency
# Contingency table
obs = np.array([[32, 15], [18,44]])
# Compute the expected frequencies and degrees of freedom
# from the given contingency table
res = chi2_contingency(obs)
print("Chi-square: ",res[0])
print("P-value: ",res[1])
```

### Correlation

```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
corr, p_value = pearsonr(x, y)
print("R-value: ", corr)
print("P-value: ", p_value)
plt.scatter(x,y,color='purple')
plt.show()
```
